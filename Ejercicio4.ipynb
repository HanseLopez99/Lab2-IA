### **Generación de una Muestra Aleatoria de una Distribución Gaussiana Multivariada**
#### Inteligencia Artificial - 2025

En este ejercicio, generaremos una muestra aleatoria de una distribución gaussiana multivariada de dimensión \( n \geq 4 \), con una media \( \mu \in \mathbb{R}^n \) y una matriz de covarianza \( \Sigma \in \mathbb{R}^{n \times n} \) especificadas por el usuario.

Posteriormente:
- Graficaremos un **pairplot** para visualizar las densidades marginales y las distribuciones bivariadas entre variables.
- Calcularemos la media y la covarianza de la muestra aleatoria.
- Compararemos los valores obtenidos con los valores teóricos especificados.


```python
# Importar librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

n = 4  # Dimensión (n >= 4)
mu = np.array([2, 5, -3, 7])  # Media en R^n
Sigma = np.array([[1.0, 0.5, 0.2, 0.1],
                  [0.5, 1.5, 0.3, 0.2],
                  [0.2, 0.3, 2.0, 0.4],
                  [0.1, 0.2, 0.4, 1.0]])


num_samples = 1000  # Número de muestras
samples = np.random.multivariate_normal(mu, Sigma, num_samples)

df = pd.DataFrame(samples, columns=[f"Var{i+1}" for i in range(n)])


df.head()
```

## **Visualización con Pairplot**
Para analizar la distribución de cada variable y sus relaciones bivariadas, generamos un **pairplot** utilizando la librería `seaborn`.


```python
# Graficar el Pairplot
sns.pairplot(df, diag_kind="kde")
plt.suptitle("Pairplot de la distribución gaussiana multivariada", y=1.02)
plt.show()
```

## **Cálculo de la Media y Covarianza Muestral**
Ahora, calcularemos la media y la matriz de covarianza de la muestra generada y las compararemos con los valores teóricos.


```python
# Cálculo de la media y covarianza de la muestra
sample_mean = np.mean(samples, axis=0)
sample_cov = np.cov(samples, rowvar=False)

print("Media Teórica:\n", mu)
print("Media Muestral:\n", sample_mean)
print("\nCovarianza Teórica:\n", Sigma)
print("Covarianza Muestral:\n", sample_cov)
```

## **Conclusión**
Si los valores de la media y la covarianza muestral son cercanos a los valores teóricos especificados, podemos concluir que la muestra generada sigue correctamente la distribución gaussiana multivariada definida.
